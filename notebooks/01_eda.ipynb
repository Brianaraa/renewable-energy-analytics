# 🔍 Exploratory Data Analysis
Global Renewable Energy Dataset

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import plotly.express as px

from src.pipeline.cleaning import run_pipeline
from src.analytics.geospatial import aggregate_by_country, generation_timeseries
from src.visualization.charts import scatter_map, choropleth_map

df = run_pipeline()
print(df.shape)
df.head()
```

```python
# Basic stats
print('Plants:', len(df))
print('Countries:', df['country'].nunique())
print('Fuel types:', df['primary_fuel'].nunique())
print('Renewable share:', df['is_renewable'].mean().round(3))
df['primary_fuel'].value_counts()
```

```python
# World map
fig = scatter_map(df.sample(10000), title='Sample of 10k Plants')
fig.show()
```

```python
# Choropleth
country_df = aggregate_by_country(df)
fig = choropleth_map(country_df)
fig.show()
```

```python
# Generation trend
ts = generation_timeseries(df)
px.line(ts, x='year', y='generation_gwh', color='primary_fuel',
        title='Generation by Fuel Type').show()
```
